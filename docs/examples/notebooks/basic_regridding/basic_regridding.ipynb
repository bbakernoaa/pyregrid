# Basic Regridding ExampleThis notebook demonstrates basic usage of PyRegrid for regridding geospatial data.

```python
import numpy as np
import xarray as xr
from pyregrid import GridRegridder

# Create sample data
lat = np.linspace(-80, 80, 40)
lon = np.linspace(-180, 180, 80)
data = np.random.rand(40, 80)

# Create DataArray
source_da = xr.DataArray(data, dims=["lat", "lon"], coords={"lat": lat, "lon": lon})
print("Source data shape:", source_da.shape)
```
